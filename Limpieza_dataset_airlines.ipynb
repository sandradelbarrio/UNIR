```python
# Import pandas library:
import pandas as pd

# Original dataset load:
df = pd.read_csv(r'airline_passenger_satisfaction.csv')

# Drop rows with 'NaN' values:
df = df.dropna()

# Drop 'ID' column:
df.drop('ID', inplace=True, axis=1)

# Count number of times each value between 1-5 repeats for the columns 8-21:
# 1 == very dissatisfied
# 2 == dissatisfied
# 3 == neutral
# 4 == satisfied
# 5 == very satisfied

# Total number of rows:

for i in range(129487):

    # Counters: 
    
    sum_very_unsatis = 0
    sum_unsatis = 0
    sum_neutral = 0
    sum_satis = 0
    sum_very_satis = 0
    val_max = 0

    
    for j in range(8, 22):

                 
        if df.iloc[i,j] == int(1):
            sum_very_unsatis = sum_very_unsatis + 1
            
            
        if df.iloc[i,j] == int(2):
            sum_unsatis = sum_unsatis + 1
            
            
        if df.iloc[i,j] == int(3):
            sum_neutral = sum_neutral + 1
            
        if df.iloc[i,j] == int(4):
            sum_satis = sum_satis + 1
            
        if df.iloc[i,j] == int(5):
            sum_very_satis = sum_very_satis + 1

    # List with counter values
            
    lista_suma = (sum_very_unsatis, sum_unsatis, sum_neutral, sum_satis, sum_very_satis)
    
    lista_suma = list(lista_suma)
    
    # Max value of the list (in case that a user has more than 1 val_max, remains the highest score)
    
    val_max = max(lista_suma)
    
    
    if sum_very_unsatis == val_max:
        df.iloc[i, 22] = "Very unsatisfied"
    
    if sum_unsatis == val_max:

        df.iloc[i, 22] = "Unsatisfied"
    
    if sum_neutral == val_max:
        df.iloc[i,22] = "Neutral"
    
    if sum_satis == val_max:
        df.iloc[i, 22] = "Satisfied"
    
    if sum_very_satis == val_max:
        df.iloc[i,22] = "Very satisfied"

df.to_csv("airline_multiclass_data_1.csv", index = True)
```
